```python
# WAP to mimic the code of capitalize method without using built in methods 

- Check if the string is empty---> return the string as it is 
- Take the first character --> convert it into the upper case letter, if it is already in the upper case letter 
- Take the remaining characters and check if any character is in uppercase ---> convert it into lowercase
- Join everything 
```

```python
def mimic_capitalize(s):

    # If the string is empty, return it 
    if len(s) == 0:
        return s
    else:
        result = '' # store the output 
        
        # Step-1:Capitalize the first letter
        fc = s[0]

        if 'a'<=fc<='z': # manually checking lowercase 
            result += chr(ord(fc)-32)
        else:
            result += fc

        # Step-2: make all the remaing character lowercase 
        for i in s[1:]:
            if 'A'<=i<='Z': # manually check for the uppercase letter 
                result += chr(ord(i)+32)
            else:
                result += i

    return result
```

```python
my_str = 'she sElls seA shellS oN tHe sea shoRe'
```

```python
mimic_capitalize(my_str)
```

```python
def mimic_split(s, delimiter = ' '):
    word = '' # store the word 
    output = [] # store the final output 
    
    for i in s:
        # check for the space 
        if i != delimiter:
            # formulate the word 
            word = i+word  # word += i
            # print(word)
        else:
            output = [word]+output
            word = ''
            # print(output)
            # print('-'*20)
            
    output = [word]+output
    return output
```
