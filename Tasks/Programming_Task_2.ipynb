### **1108. Defanging an IP Address**

Given a valid (IPv4) IP address, return a defanged version of that IP address.

A defanged IP address replaces every period "." with "[.]".

image.png

```python
def defangIPaddr(address):
    return address.replace(".","[.]")


print(defangIPaddr(address = "255.100.50.0"))
```

### **1295. Find Numbers with Even Number of Digits**

Given an array nums of integers, return how many of them contain an even number of digits.

image.png

```python
def findNumbers(nums):
    return sum(len(str(ele)) % 2 == 0 for ele in nums)

print(findNumbers(nums = [12,345,2,6,7896]))
```

### **1512. Number of Good Pairs**

Given an array of integers nums, return the number of good pairs.

A pair (i, j) is called good if nums[i] == nums[j] and i < j.

image.png

```python
def numIdenticalPairs(nums):
    gp_cnt , n = 0, len(nums)
    for i in range(n):
        for j in range(i+1,n):
            if nums[i] == nums[j] :
                gp_cnt += 1
    return gp_cnt

print(numIdenticalPairs(nums = [1,2,3,1,1,3] ))
```

### **1365. How Many Numbers Are Smaller Than the Current Number**

Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].

Return the answer in an array.



image.png

```python
from bisect import bisect_left
def smallerNumbersThanCurrent(nums):
    return [bisect_left(sorted(nums),x) for x in nums]

print(smallerNumbersThanCurrent(nums = [6,5,4,8]))
```

### **1281. Subtract the Product and Sum of Digits of an Integer**

Given an integer number n, return the difference between the product of its digits and the sum of its digits.


image.png

```python
def subtractProductAndSum(n):
    pdt , summ = 1,0
    while n>0:
        n,dig = divmod(n,10)
        pdt *= dig
        summ += dig
    return pdt - summ

print(subtractProductAndSum(n = 4421))
```

### **1486. XOR Operation in an Array**

You are given an integer n and an integer start.

Define an array *nums* where nums[i] = start + 2 * i (0-indexed) and n == nums.length.

Return the bitwise XOR of all elements of *nums*.

image.png

```python
def xorOperation(n,start):
    ans = 0
    for i in range(n):
        ans ^= start + 2*i
    return ans

print(xorOperation(n = 4,start = 3))
```
